# Task 1 : Simple Linear Regression

Predicting the score of a student depending on the amount of hours he studies.

# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Exploring the data

```python
url = 'https://bit.ly/w-data'
df = pd.read_csv(url)

df.head()
```

# Visualising the data

```python
sns.regplot(x='Hours', y='Scores', data=df, marker='p').set(title='Hours vs Percentage');
```

```python
sns.lineplot(x='Hours', y='Scores', data=df).set(title='Hours vs Percentage');
```

```python
sns.lmplot(x='Hours', y='Scores', data=df, lowess=True).set(title='Hours vs Percentage');
```

```python
sns.jointplot(x='Hours', y='Scores', data=df, kind="reg", color='g');
```

# Model Building

Importing required libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Model fitting

```python
X=np.array(df.Hours).reshape(-1,1)
y=df.Scores

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

linreg = LinearRegression().fit(X_train, y_train)
```

Checking the R-squared Scores

```python
print(f'R-squared score (training): {linreg.score(X_train, y_train):.3f}')

print(f'R-squared score (test): {linreg.score(X_test, y_test):.3f}')
```

Mean Absolute Error

```python
pred = linreg.predict(X_test)

print(f'Mean Absolute Error: {metrics.mean_absolute_error(y_test, pred)}') 
```

# Predicting score of a student studying for 9.25 hours

```python
studing_hours = np.array(9.25).reshape(1,-1)
predicted_score = linreg.predict(studing_hours)[0]

print(f'If a student studies for 9.25 hours he is likely to get a score of somewhere around {predicted_score:.1f}.')
```
